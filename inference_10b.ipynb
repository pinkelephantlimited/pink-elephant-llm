# 🐘 Pink Elephant 10B — Interactive Inference

Load the model and type your own prompts to test it.

**Note**: 10B model requires 4-bit quantization to fit on a T4 (16GB).
If you have a GPU with 24GB+ VRAM (A10G, A100, RTX 4090), remove `load_in_4bit=True` for full quality.

```python
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig
import torch

MODEL = "pinkelephantlimited/pink-elephant-10b"

quant = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_compute_dtype=torch.bfloat16,
    bnb_4bit_use_double_quant=True,
)

tokenizer = AutoTokenizer.from_pretrained(MODEL, trust_remote_code=True)
model = AutoModelForCausalLM.from_pretrained(
    MODEL,
    quantization_config=quant,
    device_map="auto",
    trust_remote_code=True,
)
print(f"✅ Loaded {MODEL} in 4-bit")
```

```python
prompt = input("Enter your prompt: ")
inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
out = model.generate(**inputs, max_new_tokens=100, do_sample=True, temperature=0.5, top_p=0.9)
print("\n" + tokenizer.decode(out[0], skip_special_tokens=True))
```

Run the cell above again to enter a new prompt.
