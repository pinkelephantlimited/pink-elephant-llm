```python
from transformers import pipeline

MODEL = "pinkelephantlimited/pink-elephant-1b"
pipe = pipeline("text-generation", model=MODEL, device_map="auto", trust_remote_code=True)
print(f"✅ Loaded {MODEL}")
```

```python
prompt = input("Enter your prompt: ")
result = pipe(prompt, max_new_tokens=80, do_sample=True, temperature=0.5)[0]["generated_text"]
print("\n" + result)
```
